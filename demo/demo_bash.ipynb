# `Hallmark` Command Line Demo

This Jupyter notebook demonstrates how `hallmark` is used in a command line.

Because all cells are shell command, it is much more natural to use a `bash` kernel.
To install a `bash` kernel to your Jupyter/python (virtual) environment, run
```bash
pip install bash_kernel
python -m bash_kernel.install
```

## Remove Old Demo Repositories

First, let's remove all old demo repositories.

````bash
rm -rf repo*
````

## Initialize a Standard (Non-Bare) `Hallmark` Repository

`Hallmark`'s commandline syntax is similar to `git`.
To create an empty (standard) `hallmark` repository, we run `hallmark init [REPO]`.

````bash
hallmark init repo1
````

Check the content of the hallmark repository.
It contains a ".hm" directory, which is itself a git repository.

````bash
ls -a repo1 repo1/.hm
````

We can `cd` into the ".hm" directory and then run `git`.
However, because `cd` is not a very good practice in Jupyter notebook,
we will use `pushd` and `popd` instead.

````bash
pushd repo1/.hm && git status && popd
````

Note, similar to `git init`, `hallmark init` does not commit your files.

## Initialize a Bare Hallmark Repository

Next, we will initialize a bare repository.
Unlikely `git`, all we need is to explicitly postfix ".hm" to the repository name.

````bash
hallmark init repo2.hm
````

`repo2.hm` is itself a `git` repository.

````bash
ls -a repo2.hm
````

````bash
pushd repo2.hm && git status && popd
````

## Add Files According to Python Format String

````bash
pushd repo1 && touch a{0,0.75,0.94}_i{0..180..30}.h5 && popd
ls repo1
````

````bash
pushd repo1 && hallmark add "a{a}_i{i}.h5" && popd
````

````bash
pushd repo1 && hallmark commit -m  "Commit test" && popd
````
