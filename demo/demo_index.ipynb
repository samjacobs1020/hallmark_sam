```python
%%bash
rm -rf repo1
rm -rf experiment
mkdir repo1
```

## 1. Initialize the hallmark repo

```python
from hallmark import Repo

repo1 = Repo.init("repo1")
```

## 2. Create sample data files on `main` with unique content

```python
%%bash
pushd repo1
for f in a{0,0.75}_i{0,30,60,90,120}.h5; do
    echo "$f" > "$f"
done
popd
ls repo1
```

## 3. Add files on `main` and commit

```python
repo = Repo.init("repo1")
pf = repo.add("a{spin}_i{inc}.h5")
print(pf)
```

## 4. Inspect the cache after `main` add

```python
%%bash
find repo1/.hm/index -type f | sort | wc -l
```

## 5. Checkout a new branch

```python
repo.checkout("experiment")
```

## 6. Create different data files on `experiment`

```python
%%bash
pushd experiment
for f in b{0.5,1.0}_i{0,45,90,135,155}.h5; do
    echo "$f" > "$f"
done
popd
ls experiment
```

## 7. Add files on `experiment`

```python
repo_exp = Repo("experiment")
pf_exp = repo_exp.add("b{spin}_i{inc}.h5")
print(pf_exp)
```
